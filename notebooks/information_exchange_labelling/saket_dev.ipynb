```python
# Read data
data_path = os.path.join(repo_path, 'data/external/gpt4_dataset')

# Read student_dominated.txt and split conversations based on "\n\n"
with open(os.path.join(data_path, 'student_dominated.txt'), 'r') as f:
    transcripts = f.read().split('\n\n')

transcripts = [x for x in transcripts if x != '']
print("Number of transcripts:", len(transcripts))

prompt_tempate_leg = """For each of the following sentences, Is the sentence information giving or information seeking? Respond with {n_utterances} lines with only "Information giving" or "Information seeking" on every line -

{utterances}
"""

prompt_tempate = """For each of the following sentences, Is the sentence information giving, information seeking or neutral? Respond with {n_utterances} lines with only "Information giving", "Information seeking" or "Neutral" on every line -

{utterances}
"""

df_data = []
for transcript in transcripts:
    # Get utterances
    utterances = transcript.split("\n")
    speakers = [x.split(":")[0] for x in utterances[:-1]]
    prompt = prompt_tempate.format(
        n_utterances=len(utterances), utterances="\n".join(utterances)
    )
    df_data.append([transcript, prompt, len(utterances), speakers])

    
df_transcripts = pd.DataFrame(df_data, columns=["transcript", "prompt", "n_utterances", "speakers"])

# Prepare prompts
utterances_per_prompt = 7

# join question with utterances n_utterances at a time
dia = [transcripts[i:i+7] for i in range(0, len(transcripts), utterances_per_prompt)]
prompt = prompt_tempate.format(n_utterances=len(utterances), utterances=utterances)
print("\nSample prompt")
print(prompts[-1])
```

```python
print(prompts[3])
```

```python
resp = agent.get_response(prompts[0])
print(agent.response_text_content(resp))

if "response" not in df_transcripts.columns:
    df_transcripts["response"] = ""

responses = []
# Get response for each prompt in df_transcripts
for i, row in df_transcripts.iterrows():
    resp = agent.get_response(row["prompt"])
    responses.append(resp)
    df_transcripts.loc[i, "response"] = agent.response_text_content(resp)
    print(df_transcripts.loc[i, "response"])
    print()
```

```python
# Split text in response into utterances
df_transcripts["response_utterances"] = df_transcripts["response"].apply(lambda x: x.split("\n"))
df_transcripts["response_n_lines"] = df_transcripts["response_utterances"].apply(lambda x: len(x))
df_transcripts["n_responses_match"] = (df_transcripts["response_n_lines"] == df_transcripts["n_utterances"])
```
